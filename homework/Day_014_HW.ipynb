```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

# 忽略警告訊息
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = './data/'
```

```python
# 讀取檔案
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版
# 資料整理 ( 'DAYS_BIRTH'全部取絕對值 )
app_train['DAYS_BIRTH'] = abs(app_train['DAYS_BIRTH'])
# 根據年齡分成不同組別 (年齡區間 - 還款與否)
data = app_train[['TARGET', 'DAYS_BIRTH', 'HOUSETYPE_MODE']] # subset
data['YEARS_BIRTH'] = data['DAYS_BIRTH'] / 365 # day-age to year-age
```

```python
# AMT_INCOME_TOTAL之分群(連續資料離散化)
data = data.dropna(axis=0,how='any')
data.index = range(len(data)) # 重新編號
unique_house_type = np.sort(data['HOUSETYPE_MODE'].unique())
print(unique_house_type)

nrows = len(unique_house_type)
ncols = nrows // 2

plt.figure(figsize=(10,30))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)

    sns.distplot(data.loc[(data['HOUSETYPE_MODE'] == unique_house_type[i]) & \
                              (data['TARGET'] == 0), "YEARS_BIRTH"], 
                 label = "TARGET = 0", hist = False)
    
    sns.distplot(data.loc[(data['HOUSETYPE_MODE'] == unique_house_type[i]) & \
                              (data['TARGET'] == 1), "YEARS_BIRTH"], 
                 label = "TARGET = 1", hist = False)
    
    plt.title(str(unique_house_type[i]))
plt.show()
```
